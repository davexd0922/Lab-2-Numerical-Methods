```python
import numpy as np

# 1. Set the step size and the point to evaluate
h = 0.01
x_val = 1.0  # Change this to whatever x value you are evaluating

# 2. Define the functions
def f1(x):
    return (x**2 + 1) / np.sin(x)

def f2(x):
    return np.sin(x) / (x**3 - 3*x**5)

def f3(x):
    numerator = np.cos(x) * np.sin(x)
    denominator = 3*x**2 + (np.cos(x + 2) / x**2)
    return numerator / denominator

# 3. Define the numerical differentiation methods
def fdd(f, x, h):
    return (f(x + h) - f(x)) / h

def bdd(f, x, h):
    return (f(x) - f(x - h)) / h

def cdd(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# 4. Execute and print results
print(f"Evaluating derivatives at x = {x_val} with Δx = {h}\n")

functions = [
    (f1, "F(x) = (x^2 + 1) / sin(x)"),
    (f2, "F(x) = sin(x) / (x^3 - 3x^5)"),
    (f3, "F(x) = (cos(x)sin(x)) / (3x^2 + cos(x+2)/x^2)")
]

for f, name in functions:
    print(f"--- {name} ---")
    print(f"FDD: {fdd(f, x_val, h):.6f}")
    print(f"BDD: {bdd(f, x_val, h):.6f}")
    print(f"CDD: {cdd(f, x_val, h):.6f}")
    print("-" * 50)
```
